```python
import csv
import agate
```

```python
tester = agate.TypeTester(force={
    'false_evidence': agate.Boolean()
})

exonerations = agate.Table.from_csv('exonerations-20150828.csv', column_types=tester)
```

```python
for row in exonerations.rows:
    print row['last_name']
```

```python
num_false_confessions = exonerations.aggregate(agate.Count('false_confession', True))

print num_false_confessions
```

```python
median_age = exonerations.aggregate(agate.Median('age'))

print median_age
```

```python
from_missouri = exonerations.aggregate(agate.Count('state', 'MO'))

print from_missouri
```

```python
exonerations.print_table()
```

```python
convicted_1990 = exonerations.where(lambda row: row['convicted'] <= 1990).where(lambda row: row['false_confession'] == True)

print convicted_1990.print_table()
```

```python
print len(convicted_1990.rows)
```

```python
with_years_in_prison = exonerations.compute([
    (agate.Change('convicted', 'exonerated'), 'years_in_prison')
])

print with_years_in_prison.select(['years_in_prison']).print_table()
```
